```python
import cv2
import os
import glob
import re

def get_sorted_frame_files(folder):
    # Find all PNG files matching frame_XXXXX.png
    files = glob.glob(os.path.join(folder, "frame_*.png"))
    # Sort by frame number
    files.sort(key=lambda x: int(re.search(r'frame_(\d+)\.png', x).group(1)))
    return files

def stitch_frames_to_video(frames_folder, output_video, fps=30):
    frame_files = get_sorted_frame_files(frames_folder)
    if not frame_files:
        print("No frames found in", frames_folder)
        return
    
    # Check if frames_folder exists, if not, create it
    if not os.path.exists(frames_folder):
        print(f"{frames_folder} does not exist. Creating it.")
        os.makedirs(frames_folder)

    # Read first frame to get size
    frame = cv2.imread(frame_files[0])
    height, width, layers = frame.shape

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # or 'XVID'
    video = cv2.VideoWriter(output_video, fourcc, fps, (width, height))

    for fname in frame_files:
        img = cv2.imread(fname)
        if img is None:
            print(f"Warning: Could not read {fname}")
            continue
        video.write(img)

    video.release()
    print(f"Video saved as {output_video}")

if __name__ == "__main__":
    # Change these paths as needed
    frames_folder = "SignalDataFrames/ZF.FishSignalData07_20250530_141502"
    output_video = "ZF.Fish07Signaldata_video.mp4"
    fps = 30

    stitch_frames_to_video(frames_folder, output_video, fps)
```
